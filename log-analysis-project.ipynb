```python
import numpy as np
import re
from datetime import datetime

```

```python
#Reading the logs file
with open("access.log","r") as log_file:
    content=log_file.read()  
```

```python

logs=[] #List of logs
for log in re.finditer(r"(?P<HostName>[\w\.-]+\.?[a-zA-z0-9]+) - - (?P<Timestamp>\[\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2} -\d{4}\]) \"(?P<Request>[A-Z]{3}\s.+\") (?P<Status>\d{3}) (?P<ResponseSize>\d+)",content):
    print(log.groupdict())
    logs.append(log.groupdict())
```

```python
#Calculate The Average Requests per Hour
liRe=[] #List of timestamps
res=[] #List of responses size
hosts=[] #List of Hosts
most_host={} # Dictionary of hosts with their Requests frequency number.
for log in logs:
    liRe.append(log["Timestamp"].strip("[]"))
    res.append(int(log["ResponseSize"]))
    hosts.append(log["HostName"])

timestamps =np.array(liRe)  


times = np.array([datetime.strptime(ts, "%d/%b/%Y:%H:%M:%S %z") for ts in timestamps])
first=times.min()
last=times.max()
total_hours=(last-first).total_seconds() / 3600
avg_requests = len(times) / total_hours
print("Average Requests per Hour:",avg_requests)

    
    
```

```python
#calculate The Percentage of successful status codes
success = 0
for code in logs:
    if code["Status"].startswith("2"):
        success += 1
percentage = (success / len(logs)) * 100
print(f"percentage of successful status codes: {percentage:.2f}%")
```

```python
#Calculate The Average Responses size
response=np.array(res)
avg_response=response.mean()
print("Average Response size:",avg_response)
```

```python
#Evaluate The most IP address that submitted a request

for host in hosts:
    if host not in most_host:
        most_host[host]=1
    else:
        most_host[host]+=1
maxHostn =list(most_host.keys())[0]

for host in most_host:
    if most_host[maxHostn] < most_host[host]:
        maxHostn=host                        
print("The most IP address that submitted a request: ",maxHostn,"with",most_host[maxHostn],"times")
```
